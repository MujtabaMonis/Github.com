```python
import pandas as pd
```

```python
df=pd.read_csv('Salaries.csv')
```

```python
df
```

```python
# Q1
df = df[~((df['EmployeeName'] == 'Not provided') & (df['JobTitle'] == 'Not provided'))]
```

```python
# Q2
df.loc[:, ['BasePay','OvertimePay','OtherPay','Benefits']] = (
    df[['BasePay','OvertimePay','OtherPay','Benefits']].replace('Not Provided', 0)
)
```

```python
# Q3
df.loc[:, ['BasePay','OvertimePay','OtherPay','Benefits']] = (
    df[['BasePay','OvertimePay','OtherPay','Benefits']].astype(float)
)
```

```python
# Q4
df[df['BasePay'] > 200000][['EmployeeName','BasePay']]
```

```python
# Q5
df[df['JobTitle'].str.contains('Manager', case=False, na=False)]
```
